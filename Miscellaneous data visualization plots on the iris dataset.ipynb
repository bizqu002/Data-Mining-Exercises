```python
###Introduction to Data Mining by Tan, Steinbach & Kumar
###Chapter 3 Problem 1
###Author: Brenda S. Izquierdo

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import numpy as np
import pandas as pd
from sklearn import datasets
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates, andrews_curves, radviz
sns.set(color_codes = True)

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
labels = ["Sepal Lenght", "Sepal Width", "Petal Length", "Petal Width"]
species = ["Setosa", "Versicolour", "Virginica"]
target = iris.target
target_df = pd.DataFrame(iris.target)
iris_pair_df = iris_df
species_dict = {0: species[0], 1: species[1], 2: species[2]}
iris_pair_df["species"] = target_df
iris_pair_df["species"] = iris_pair_df["species"].map(species_dict )
iris_pair_df.head()

```

```python
#Histogram
ax = sns.distplot(iris_df["petal length (cm)"], bins = 10)
ax.set_title("10 bin histogram of Iris Dataset (petal length)")
plt.show()
```

```python
#2D Joint Plot
sns.jointplot("sepal length (cm)", "sepal width (cm)", iris_df[["sepal length (cm)", "sepal width (cm)"]], kind = "hex", color="darkgreen")
plt.title("2D Joint Plot of Iris Dataset, sepal length vs sepal width")
plt.show()
```

```python
#Box Plot
ax = sns.boxplot(data = iris_df)
plt.title("Box Plot for Iris Dataset")
plt.show()
```

```python
#Pie Chart
colors = ['#FEBFB3', '#E1396C', '#96D38C', '#D0F9B1']
plt.title("Distribution of the types of Iris flowers")
plt.pie(np.bincount(target), labels=species, autopct="%.1f%%", colors=colors)
plt.axis("equal")
plt.show()
```

```python
#Pair Plot
sns.pairplot(iris_df, hue="species", size=1.5)
plt.show()
```

```python
#Scatter Plot
plt.title("Scatter Plot, petal length vs petal width ")
plt.scatter(x = iris_df["petal length (cm)"], y = iris_df["petal width (cm)"], c='#E1396C')
plt.show()
```

```python
#Parallel Coordinates
parallel_coordinates(iris_pair_df, "species", color=colors)
plt.title("Parallel Coordinates visualization for Iris Dataset")
plt.show()
```

```python
#Andrews Curves
andrews_curves(iris_pair_df, "species")
plt.title("Andrews Curves visualization for Iris Dataset")
plt.show()
```

```python

#Autocorrelation Plot
radviz(iris_pair_df, "species")
plt.title("RadViz visualization for Iris Dataset")
plt.show()
```
